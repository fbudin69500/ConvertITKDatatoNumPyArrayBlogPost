## 2. Import packages

```python
%matplotlib inline
import itk
import matplotlib.pyplot as plt
```

## 3. Read images using ITK

```python
filename = "ITK-Python-Wrapping-Take-2-Edited.png"

ImageType = itk.Image[itk.UC,2]
reader = itk.ImageFileReader[ImageType].New(FileName=filename)
reader.Update()

img = reader.GetOutput()
```

## 4. Convert ITK image to a NumPy ArrayView

```python
arr = itk.GetArrayViewFromImage(img)
```

## 5. Display image

```python
plt.figure(figsize=(8,8))
imgplot = plt.imshow(arr, cmap=plt.cm.hot)
```

## 6. Print raw array

```python
print(arr)
```

## 7. Save array

```python
plt.imsave('ConvertITKDataToNumPyArray.png', arr, cmap=plt.cm.hot)
```
